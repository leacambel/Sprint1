```python
# Import Libraries
# Pandas so you can read csv files
import pandas as pd
# Bar chart Race
import bar_chart_race as bcr
```

```python
# input your csv to a variable
df_mmda = pd.read_csv('mmda.csv')
```

```python
# See the columns
df_mmda
```

```python
# Check Values of this column = Lanes Block
df_mmda['Lanes_Blocked'].value_counts()
```

```python
# I only need these 3 Columns so I filtered them
data_mmda = df_mmda[['Date','Lanes_Blocked','Involved']]
```

```python
# See how many values per column
data_mmda.count()
```

```python
data_mmda['Involved'].value_counts()
```

```python
data_mmda['Involved'].value_counts()
```

```python
# Mali spelling ng paranaque
data_mmda = data_mmda.replace(['ParaÃ±aque'], ['Parañaque']) 
data_mmda['City'].value_counts()
```

```python
data_mmda
```

```python
# Convert this column to date
data_mmda['Date'] = pd.to_datetime(data_mmda.Date)
```

```python
# Got year and month
data_mmda['month_year'] = data_mmda['Date'].dt.to_period('M')
```

```python
data_mmda
```

```python
# Pivot table like in excel check (para columns ay city)
mmda_piv = data_mmda.pivot_table(index='month_year', columns='Involved', values='Lanes_Blocked', aggfunc='sum')
mmda_piv.head()
```

```python
# Fill None values with zeroes
mmda_piv = mmda_piv.fillna(0)
```

```python
# Cumulate the date 
mmda_data = mmda_piv.cumsum()
mmda_data
```

```python
 [1,1 , 1, 1, 1 ]
```

```python
bcr.bar_chart_race(df=mmda_data,
                   n_bars=15,
                   title='Total Number Lanes Blocked',
                  bar_size=.7,
                   sort='desc',
                   period_length=700,
                  fixed_max=True,
                  orientation='h')
```

```python
# Import Libraries
# Pandas so you can read csv files
import pandas as pd
# Bar chart Race
import bar_chart_race as bcr
```

```python
# https://www.kaggle.com/grosvenpaul/dengue-cases-in-the-philippines
# thank you to the Department of Health of the Philippines for providing the raw data
df_dengue = pd.read_csv('denguecases.csv')
```

```python
# Year column to datetime format
df_dengue.Year = pd.to_datetime(df_dengue.Year, format='%Y')
```

```python
deng_piv = df_dengue.pivot_table(index='Year', columns='Region', values='Dengue_Cases', aggfunc='sum')
```

```python
df_dengue
```

```python
deng_piv.head()
```

```python
dengue_data = deng_piv.cumsum()
```

```python
dengue_data.head(10)
```

```python
bcr.bar_chart_race(df=dengue_data,
                   n_bars=17,
                   title='Number of dengue cases per 100,000 population (2008 to 2016)',
                  bar_size=.7,
                   sort='desc',
                   period_length=1000,
                  fixed_max=True,
                  orientation='h',
                  perpendicular_bar_func='median',
                  period_fmt='%Y')
```
